### While-loop

```python
guess_me = 7
number = 1

while True:
  if number < guess_me:
    print("too low")
  elif number == guess_me:
    print("found it!")
    break
  else:
    print("oops")
    break

  number += 1
```

```python
number
```

### for-loop

```python
for number in range(1, 10):
  if number < guess_me:
    print("too low")
  elif number == guess_me:
    print("found it!")
    break
  else:
    print("oops")
    break
```
