```python
(
    flights
    .nlargest(
        5, 'dep_delay',
        keep = 'all'
    )
    .merge(airlines,
           on = 'carrier',
           how = 'left')
    [['name','dep_delay']]
)
```

### Q5
Consider the following two airlines:

- Delta Air Lines Inc. (DL)
- United Air Lines Inc. (UA)

Determine which airline has a higher proportion of flights with a dep_delay greater than 30 minutes.

```python
import numpy as np
```

```python
q5 = (
    flights
    .query(
       'carrier.isin(["DL", "UA"])'
    )
    [['dep_delay', 'carrier']]
)

q5['cat'] = np.where(
   q5['dep_delay'] > 30, True, False
)

q5_final = (
    q5[['carrier', 'cat']]
    .value_counts()
    .reset_index()
)

q5_final
```

```python
DL_num = q5_final.query('carrier == "DL" & cat == True')['count'].values
DL_denom = q5_final.query('carrier == "DL"')['count'].sum()
```

```python
DL_num
```

```python
DL_denom
```

```python
DL_num / DL_denom
```

```python
UA_num = q5_final.query('carrier == "UA" & cat == True')['count'].values
UA_denom = q5_final.query('carrier == "UA"')['count'].sum()
```

```python
UA_num / UA_denom
```

```python
DL_num / DL_denom < UA_num / UA_denom
```
