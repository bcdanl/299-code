```python
import pandas as pd
from google.colab import data_table
data_table.enable_dataframe_formatter()

df_wide = pd.DataFrame({
    'Weekday': ['Tuesday', 'Wednesday'],
    'Miami': [80, 83],
    'Rochester': [57, 62],
    'St. Louis': [71, 75]
})

df_long = pd.DataFrame({
    'Weekday': ['Tuesday', 'Wednesday', 'Tuesday', 'Wednesday', 'Tuesday', 'Wednesday'],
    'City': ['Miami', 'Miami', 'Rochester', 'Rochester', 'St. Louis', 'St. Louis'],
    'Temperature': [80, 83, 57, 62, 71, 75]
})
```

# melt() - to make dataframe longer

```python
df_wide_to_long = (
    df_wide
    .melt()
)

df_wide_to_long
```

```python
df_wide_to_long = (
    df_wide
    .melt(id_vars = "Weekday")
)

df_wide_to_long
```

```python
df_wide_to_long = (
    df_wide
    .melt(id_vars = "Weekday",
          var_name= "City",
          value_name= "Temperature")
)

df_wide_to_long
```

```python
df_long
```

# pivot() - to make dataframe wider

```python
df_long
```

```python
df_long_to_wide = (
    df_long
    .pivot(
        index = "Weekday",
        columns = "City",
        values = "Temperature"
    )
)

df_long_to_wide
```

```python
df_long_to_wide = (
    df_long
    .pivot(
        index = "Weekday",
        columns = "City",
        values = "Temperature"
    )
    .reset_index()
)

df_long_to_wide
```

```python
dict_data = {"Name": ["Donna", "Donna", "Mike", "Mike"],
             "Department": ["ECON", "DANL", "ECON", "DANL"],
             "2018": [1, 2, 3, 1],
             "2019": [2, 3, 4, 2],
             "2020": [5, 1, 2, 2]}
df = pd.DataFrame(dict_data)

df_longer = df.melt(id_vars=["Name", "Department"],
                    var_name="Year",
                    value_name="Number of Courses")
```

```python
df
```

```python
df_longer
```

```python
df_wider = (
    df_longer
    .pivot(
        index = ['Name', 'Department'],
        columns = 'Year',
        values = 'Number of Courses'
    )
    .reset_index()
    .sort_values(['Name', 'Department'], ascending= [True, False])
)

df_wider
```

# merge() - to join related dataframes

```python
x = pd.DataFrame({
    'key': [1, 2, 3],
    'val_x': ['x1', 'x2', 'x3']
})

y = pd.DataFrame({
    'key': [1, 2, 4],
    'val_y': ['y1', 'y2', 'y3']
})
```

```python
x
```

```python
y
```

```python
x.merge(y, on = "key")
```

```python
x.merge(y, on = "key", how = 'inner')
```

```python
pd.merge(x, y, on = "key")
```

```python
pd.merge(x, y, on = "key", how = 'inner')
```

```python
x.merge(y, on = 'key', how = 'left')
```

```python
x.merge(y, on = 'key', how = 'right')
```

```python
x.merge(y, on = 'key', how = 'outer')
```

# duplicate keys

```python
x = pd.DataFrame({
    'key':[1, 2, 2, 3],
    'val_x':['x1', 'x2', 'x3', 'x4']})

y = pd.DataFrame({
    'key':[1, 2],
    'val_y':['y1', 'y2'] })
```

```python
x
```

```python
y
```

```python
one_to_many = x.merge(y, on='key',
                         how='left')

one_to_many
```

```python
x = pd.DataFrame({
  'key':[1, 2, 2, 3],
  'val_x':['x1','x2','x3','x4']})
```

```python
y = pd.DataFrame({
  'key': [1, 2, 2, 3],
  'val_y': ['y1', 'y2', 'y3', 'y4'] })
```

```python
x
```

```python
y
```

```python
many_to_many = x.merge(y, on='key',
                          how='left')

many_to_many
```

# Defining the key columns

```python
x = pd.DataFrame({
  'key_x': [1, 2, 3],
  'val_x': ['x1', 'x2', 'x3']
})
```

```python
y = pd.DataFrame({
  'key_y': [1, 2],
  'val_y': ['y1', 'y2'] })
```

```python
x
```

```python
y
```

```python
x.merge(y, on = 'key', how = 'left')
```

```python
x.merge(y,
        left_on = 'key_x',
        right_on = 'key_y',
        how = 'left')
```
